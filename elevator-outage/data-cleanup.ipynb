```python

import pandas as pd
import numpy as np

#read csv file
df = pd.read_csv('data/data-cleanup-io/2023_subway_censusvar_agg.csv')
df2 = pd.read_csv('data/data-cleanup-io/elevator-avail-2023.csv')
```

```python
# show me the column names in df2
print(df2.columns)
```

```python
print (df.columns)
```

```python
#rename columns
df2.rename(columns={'Equipment Code':'Common Name'}, inplace=True)
```

```python


# Perform a left join to merge the DataFrames on the "Common Name" column
df_updated = pd.merge(df, df2[['Common Name', 'Time Since Major Improvement']], on='Common Name', how='left')



```

```python
# Do not run this cell; the columns have already been dropped

# Drop rows where "Common Name" or "Time Since Major Improvement" is NaN before the operation
df_clean = df_updated.dropna(subset=['Common Name', 'Time Since Major Improvement'])

# Then, proceed with grouping and finding the index of the max "Time Since Major Improvement"
idx = df_clean.groupby('Common Name')['Time Since Major Improvement'].idxmax()

# Select rows by the indices found
df_max_improvement = df_clean.loc[idx]

# df_max_improvement now contains rows of the highest "Time Since Major Improvement" for each "Common Name"
```

```python
#write to csv
df_max_improvement.to_csv('data/data-cleanup-io/2023_subway_merged2.csv', index=False)
```

```python
#read csv file 
df = pd.read_csv('data/data-cleanup-io/average-weekday-rider-2021.csv')
```

```python
import pandas as pd


# Step 2: Extract data within brackets in the "Station Complex Name" column using a more specific regex pattern
# Adjusted regex pattern to capture the content within parentheses, assuming the actual delimiter is a parenthesis and not a bracket
df['Daytime Routes'] = df['Station Complex Name'].str.extract('\((.*?)\)')

# Step 3: Clean the extracted data - remove commas if they are not needed
df['Daytime Routes'] = df['Daytime Routes'].str.replace(',', ' ')

# Optionally, Step 4: Remove the data within brackets from the original "Station Complex Name" column
df['Station Complex Name'] = df['Station Complex Name'].str.replace('\s*\([^)]*\)\s*', '', regex=True)

# Show the result
print(df[['Station Complex Name', '2021', 'Daytime Routes']].head())
```

```python
#writing to csv
df.to_csv('data/data-cleanup-io/average-weekday-rider-2021 new.csv', index=False)
```

```python
import pandas as pd
import numpy as np
```

```python
# Read the CSV files
df1 = pd.read_csv('elevator_final.csv')
print(df1.columns)
```

```python
# Station MRN_x, ADA Notes, Time Since Major Improvement_x, Month, Total Outages, 'Scheduled Outages', 'Unscheduled Outages', 'Entrapments', 'Station Name', 'Station MRN_y', 'Station Complex Name', 'Station Complex MRN',
# drop the columns
df1.drop(['Station MRN_x', 'ADA Notes', 'Time Since Major Improvement_x', 'Month', 'Total Outages', 'Scheduled Outages', 'Unscheduled Outages', 'Entrapments', 'Station MRN_y', 'Station Complex Name', 'Station Complex MRN'], axis=1, inplace=True)
print(df1.columns)
```

```python
#dtypes of columns
print(df1.dtypes)
```

```python

# Function to count characters excluding spaces
def count_characters(s):#
    return len(s.replace(' ', ''))


df1['Route count'] = df1['Daytime Routes'].apply(count_characters)
df1['Major Interchange'] = (df1['Route count'] >= 3).astype(int)
#print column names
print(df1.columns)

```

```python
# Group the data by 'Station Name' and aggregate both the sum, average, and median of 'outage_count'
df1_grouped = df1.groupby('Station Name')['outage_count'].agg(
    total_outage_count='sum',
    average_outage_count='mean',
    median_outage_count='median'
).reset_index()

# Display the resulting DataFrame
df1_grouped

```

```python
# add amb_pop_estimate column from df1 to df1_grouped, merge on Stop Name
df1_grouped = pd.merge(df1_grouped, df1[['Station Name', 'amb_pop_estimate','Ridership','Time Since Major Improvement','med_earn_estimate','Route count','Major Interchange','population_estimate']], on='Station Name', how='left')
```

```python
df1_grouped
```

```python
df1_grouped = df1_grouped.drop_duplicates(subset='Station Name', keep='first')
df1_grouped
```

```python
#write to csv
df1_grouped.to_csv('data/data-cleanup-io/df1_grouped.csv', index=False)
```
